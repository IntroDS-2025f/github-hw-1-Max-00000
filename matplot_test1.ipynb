````python
# ‰Ω†ÁöÑÁ®ãÂºè

import numpy as np
import matplotlib.pyplot as plt

# Âõ∫ÂÆö‰∫ÇÊï∏Á®ÆÂ≠ê‰ª•‰æøÁµêÊûúÂèØÈáçÁèæ
np.random.seed(768)

n = 200

x = np.random.normal(loc=2.0, scale=1.0, size=n)
y = x * 0.5 + np.random.normal(scale=0.8, size=n)
size_var = np.abs(np.random.normal(loc=20, scale=10, size=n))
size_scalar = (size_var - size_var.min()) / (size_var.max() - size_var.min() + 1e-9) * 490 + 10

fig, ax = plt.subplots(
    nrows=1,
    ncols=1,
    layout='constrained'
)

ax.set(
    title='Scatter with Colormap',
    xlim=(-2, 6),
    ylim=(-2, 4)
)
ax.grid(visible=True)

sc = ax.scatter(x, y, c=y, s=size_scalar, alpha=0.75, edgecolors='black', linewidths=0.5)

cbar = fig.colorbar(sc, ax=ax)

plt.show()
````
